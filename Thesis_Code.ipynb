```python
Cytot_sinQ_Case1 = (16*exp1['A/D']*math.pi*zeta1*m_osc_exp1*f_n_v_exp1*exp1['f*']*1.226)/(ro*D*L*(exp1['U (m/s)']**2))
Cytot_sinQ_Case2 = (16*exp2['A/D']*math.pi*zeta2*m_osc_exp2*f_n_v_exp2*exp2['f*']*1.1797)/(ro*D*L*(exp2['U (m/s)']**2))
Cytot_sinQ_Case3 = (16*exp3['A/D']*math.pi*zeta3*m_osc_exp3*f_n_v_exp3*exp3['f*']*1.142)/(ro*D*L*(exp3['U (m/s)']**2))
Cytot_sinQ_Case4 = (16*exp4['A/D']*math.pi*zeta4*m_osc_exp4*f_n_v_exp4*exp4['f*']*1.209)/(ro*D*L*(exp4['U (m/s)']**2))
Cytot_sinQ_Case5 = (16*exp5['A/D']*math.pi*zeta5*m_osc_exp5*f_n_v_exp5*exp5['f*']*1.153)/(ro*D*L*(exp5['U (m/s)']**2))
```
